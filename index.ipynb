Loading data and importing pandas for  
 Aviation Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Aviation_Data.csv')
df.head()
```

```python
# show first 5 rows of the dataframe
df.head()
```

```python
# show last 5 rows of the dataframe
df.tail()
```

```python
# summary of the dataframe
df.info()
```

```python
# sum of null values in each column
df.isnull().sum()
```

```python
# create new column for percentage of null values in each column
null_percentage =  (df.isnull().sum()/len(df)*100).round(2)
null_percentage
```

```python
# list of null values less than 10%
column_10 = [key for key,value in null_percentage.items() if value <10]
column_10
```

```python
# drop values with less than 10% null values
df.dropna(subset= column_10, inplace= True)
```

```python
# checking if null values are dropped
df.isnull().sum()
```

```python
# dropping less important columns
df.drop(columns=['Latitude', 'Longitude', 'Publication.Date', 'Airport.Name', 'Airport.Code', 'FAR.Description', 'Schedule', 'Air.carrier'], inplace=True)
```

Fill null values

```python
# fill null values in total facial injuries with 0
df['Total.Fatal.Injuries'].fillna(0,inplace=True)
# fill null values in total minor injuries with 0
df['Total.Minor.Injuries'].fillna(0,inplace=True)
# fill null values in total serious  injuries with 0
df['Total.Serious.Injuries'].fillna(0,inplace=True)
# fill null values with unknown
df['Broad.phase.of.flight'].fillna('Unknown', inplace= True)
```

```python
# normalize the make column first
df['Make'] = df['Make'].str.upper()
```

```python
# create a function that 
def fill_nan_aircraft_category(df):
    df['Aircraft.Category'] = df.groupby(['Make'])['Aircraft.Category'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
    return df
fill_nan_aircraft_category(df)
```

```python
# checking again for null values
df.isnull().sum()
```

```python
# checking for duplicates
df.duplicated().sum()
```

Accident trends over time

```python
# convert event date column to date time format inorder to extract year from date
df['Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Date'].dt.year
```

```python
# Counts number of accidents per year
accidents_per_year = df['Year'].value_counts().sort_index()
accidents_per_year
```

```python
# Plot a lineplot to show the trend of accidents over time
plt.figure(figsize=(12, 6))
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

- It looks like there was an increase of accident in 1982
- After a high rate of accidents, it could mean that there was an increase in accident mitigation as accidents reduced progressively over time after 1982.
- This mitigations could be attributed from i.e number of engines, engine type, broad phase of light

Accident Trends by Aircraft Make


```python
# Top 10 aircraft makes with the most accidents
top_makes = df['Make'].value_counts().nlargest(10).index
top_makes
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=df[df['Make'].isin(top_makes)], x='Make', order=top_makes)
plt.title('Top 10 Makes with Most Accidents')
plt.xlabel('Aircraft Make')
plt.ylabel('Nb of Accidents')
plt.xticks(rotation=60)
plt.yscale('log')
plt.show()
```

We are creating a countplot to show which aircraft make has the most accidents over time.

1. Cessna has had the highest number of accidents followed by Piper.
2. Among top 10 aircraft make Boeing has had lowest accidents over time

Injuries by Phase of Flight

```python
phase_of_flight_data = df.groupby('Broad.phase.of.flight').agg(
    total_incidents=('Event.Id', 'count'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum')
).reset_index()
phase_of_flight_data
```

```python
# Plot the size of the data
plt.figure(figsize=(14, 8))

# Total incidents occured by flight
plt.subplot(2, 1, 1)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_incidents'], color='b')
plt.title('Total Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(True)

# Total injuries by phase of flight
plt.subplot(2, 1, 2)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_fatal_injuries'], color='b', alpha=0.5, label='Fatal Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_serious_injuries'], color='yellow', alpha=0.5, label='Serious Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_minor_injuries'], color='r', alpha=0.5, label='Minor Injuries')
plt.title('Total Injuries by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Injuries')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

I have created Two subplots to focus on. This helps understand the distribution of incidents and injuries across different phases of flight.

1. The first subplot shows the total number of incidents{{fatal+ serious + minor injuries} by the phase of flight.
2. The second subplot shows the total number of fatal, serious, and minor injuries by phase of flight.
